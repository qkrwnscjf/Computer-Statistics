```python
#강의 자료에서 만들어진 country4 dataset 을 이용하여, 문자열이 'NaN' 인 경우에는 'unknown' 으로,

#숫자가 'NaN' 인 경우에는 -9999 로 대체하여, 완성된 데이터셋을 new_country로 저장하시오. 

import numpy as np
import pandas as pd
```

```python
country  = pd.read_csv("country.csv")
country.head()
```

```python
country2 = pd.read_csv("country.csv", index_col = 0)
country2.head()
```

```python
country2['density'] = country2['population']/country2['area']
country2.head()
```

```python
new_value = pd.DataFrame(index = ['FR'], data = [('France', 265449, 'Paris', 126793004, 34567)],
                        columns = ['country', 'area', 'capital', 'population', 'density'])

country4 = pd.concat([country2, new_value])
#country4 = country2.append(new_value) - 오류 발생 
country4
```

```python
country4['capital'].fillna('Unknown', inplace = True)
#country4
country4['area'].fillna(-9999, inplace = True)
#country4
country4['density'].fillna(-9999, inplace = True)
#country4
new_country = country4
new_country
```
